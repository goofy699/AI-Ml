Mathematical Operations on List

```python
def s(l):
  return sum(l)
def a(l):
  return sum(l)/len(l)
def mx(l):
  return max(l)
def mn(l):
  return min(l)

try:
    l = list(map(float,input("nums: ").split()))
    o = input("1 sum 2 avg 3 max 4 min: ")

    if o=="1":
        print(s(l))
    elif o=="2":
        print(a(l))
    elif o=="3":
        print(mx(l))
    elif o=="4":
        print(mn(l))

except:
    print("err")
```

```python
#list manipulation
#every other
def eo(l):
    return l[::2]

l=[1,2,3,4,5,6]
print("input:",l)
print("output:",eo(l))
```

```python
#slice
def sub(l,s,e):
    return l[s:e+1]

l=[1,2,3,4,5,6]
print("input:",l)
print("output:",sub(l,2,4))
```

```python
#reverse list
def rev(l):
    return l[::-1]

l=[1,2,3,4,5]
print("input:",l)
print("output:",rev(l))
```

```python
# remove first and last
def rm(l):
    return l[1:-1]

l=[1,2,3,4,5]
print("input:",l)
print("output:",rm(l))
```

```python
#get first n elements
def fn(l,n):
    return l[:n]

l=[1,2,3,4,5]
print("input:",l)
print("output:",fn(l,3))
```

```python
#get last n
def ln(l,n):
    return l[-n:]

l=[1,2,3,4,5]
print("input:",l)
print("output:",ln(l,2))
```

```python
#reverse skip
def rs(l):
    return l[-2::-2]

l=[1,2,3,4,5,6]
print("input:",l)
print("output:",rs(l))
```

```python
#flatten a nested list
def flat(l):
    r=[]
    for i in l:
        r+=i
    return r

l=[[1,2],[3,4],[5]]
print("input:",l)
print("output:",flat(l))
```

```python
#access nested element
def acc(l,idx):
    x=l
    for i in idx:
        x=x[i]
    return x

l=[[1,2,3],[4,5,6],[7,8,9]]
i=[1,2]

print("input:",l)
print("indices:",i)
print("output:",acc(l,i))
```

```python
#sum of nested list
def sn(l):
    t=0
    for i in l:
        if type(i)==list:
            t+=sn(i)
        else:
            t+=i
    return t

l=[[1,2],[3,[4,5]],6]

print("input:",l)
print("output:",sn(l))
```

```python
#remove specific element
def rem(l,e):
    r=[]
    for i in l:
        if type(i)==list:
            r.append(rem(i,e))
        elif i!=e:
            r.append(i)
    return r

l=[[1,2],[3,2],[4,5]]

print("input:",l)
print("remove:",2)
print("output:",rem(l,2))
```

```python
#find max in nested list
def fm(l):
    m=-999999
    for i in l:
        if type(i)==list:
            m=max(m,fm(i))
        else:
            m=max(m,i)
    return m

l=[[1,2],[3,[4,5]],6]

print("input:",l)
print("output:",fm(l))
```

```python
#count occurences
def co(l,e):
    c=0
    for i in l:
        if type(i)==list:
            c+=co(i,e)
        elif i==e:
            c+=1
    return c

l=[[1,2],[2,3],[2,4]]

print("input:",l)
print("count:",2)
print("output:",co(l,2))
```

```python
#deep flatten
def df(l):
    r=[]
    for i in l:
        if type(i)==list:
            r+=df(i)
        else:
            r.append(i)
    return r

l=[[[1,2],[3,4]],[[5,6],[7,8]]]

print("input:",l)
print("output:",df(l))
```

```python
#average of nested list
def df(l):
    r=[]
    for i in l:
        if type(i)==list:
            r+=df(i)
        else:
            r.append(i)
    return r

def avg(l):
    f=df(l)
    return sum(f)/len(f)

l=[[1,2],[3,4],[5,6]]

print("input:",l)
print("output:",avg(l))
```

```python
#matrix operations
#Prove A.A−1 = I.

A=np.array([[3,4],[7,8]])
B=np.array([[5,3],[2,1]])

ai=np.linalg.inv(A)
i=np.dot(A,ai)

print("A inverse =\n",ai)
print("A.A^-1 =\n",i)

#Prove AB ̸= BA.
A=np.array([[3,4],[7,8]])
B=np.array([[5,3],[2,1]])

ab=np.dot(A,B)
ba=np.dot(B,A)

print("AB =\n",ab)
print("BA =\n",ba)
print("AB != BA =",not np.array_equal(ab,ba))

#Prove (AB)^T = B^T.A^T
A=np.array([[3,4],[7,8]])
B=np.array([[5,3],[2,1]])

ab=np.dot(A,B)

l=ab.T
r=np.dot(B.T,A.T)

print("(AB)^T =\n",l)
print("B^T A^T =\n",r)
print("equal =",np.array_equal(l,r))

```

```python
#Solve the following system of Linear equation using Inverse Methods.
#2x − 3y + z = −1
#x − y + 2z = −3
#3x + y − z = 9


A=np.array([[2,-3,1],
            [1,-1,2],
            [3,1,-1]])

B=np.array([[-1],
            [-3],
            [9]])

print("A =\n",A)
print("B =\n",B)

#find inverse of A
ai=np.linalg.inv(A)
print("A inverse=\n",ai)

#X = A^-1 B
X=np.dot(ai,B)

print("solution X=\n",X)
```

```python
A=np.array([[2,-3,1],
            [1,-1,2],
            [3,1,-1]])

B=np.array([[-1],
            [-3],
            [9]])

#direct solution
X=np.linalg.inv(A)@B

print("solution=\n",X)
```

```python
#Dot product
n=1000000
a=list(range(n))
b=list(range(n))

s=time.time()

dot=0
for i in range(n):
    dot+=a[i]*b[i]

e=time.time()

print("python dot product =",dot)
print("time =",e-s)

a=np.arange(n)
b=np.arange(n)

s_np=time.time()

dot=np.dot(a,b)

e_np=time.time()

print("numpy dot product =",dot)
print("time =",e_np-s_np)
```

```python
#matrix multiplication
#python lists

n=1000

A=[[1]*n for _ in range(n)]
B=[[1]*n for _ in range(n)]
C=[[0]*n for _ in range(n)]

s=time.time()

for i in range(n):
    for j in range(n):
        for k in range(n):
            C[i][j]+=A[i][k]*B[k][j]

e=time.time()

print("python matrix mult time =",e-s)
```
